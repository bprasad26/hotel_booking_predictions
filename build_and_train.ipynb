## Data Preprocessing for ML 

```python
X_train = train.drop('is_canceled', axis=1)
y_train = train['is_canceled']
X_test = test.drop('is_canceled', axis=1)
y_test = test['is_canceled']
```

```python
X_train.head()
```

```python
num_cols = X_train.select_dtypes(include=['number']).columns.to_list()
```

```python
def num_to_cat(df, num_col_list, n_unique_val):
    """This function takes a pandas dataframe, a list of numerical columns
    and create a list of columns that needs to be converted to categorical column if 
    it is less than or equal to n_unique_val."""
    
    # columns that needs to converted
    cols_to_convert = []
    for col in num_col_list:
        unique_val = df[col].nunique()
        print(col, unique_val)
        if unique_val <= n_unique_val:
            cols_to_convert.append(col)
    return cols_to_convert
        
```

```python
a_list = num_to_cat(X_train, num_cols, 25)
```

```python
a_list
```

```python
# convert these columns to categorical
for col in a_list:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')
```

```python
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns.to_list()
cat_cols = [col for col in cat_cols if col != 'country']
```

```python
cat_cols
```

### Create Pipeline 

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Numerical columns pipeline
num_pipeline = Pipeline([('num_imputer', SimpleImputer(strategy='median')),
                        ('std_scaler', StandardScaler()),])

# categorical columns pipeline
cat_pipeline = Pipeline([('cat_imputer', SimpleImputer(strategy='most_frequent')),
                        ('one_hot_encoder', OneHotEncoder(sparse=False, handle_unknown='ignore')),])

# full pipeline
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols),
])
```

```python
X_train = full_pipeline.fit_transform(X_train)
```

```python
X_test = full_pipeline.transform(X_test)
```

```python
full_pipeline.named_transformers_
```

```python
# Get the list of categories generated by the one-hot-encoder
ohe_categories = full_pipeline.named_transformers_['cat']['one_hot_encoder'].categories_
ohe_categories
```

```python
# Create nice names for our one hot encoded features
new_ohe_features = [f"{col}__{val}" for col, vals in zip(cat_cols, ohe_categories) for val in vals]
new_ohe_features
```

```python
# Create a new list with all names of features
all_features = num_cols + new_ohe_features
all_features
```

```python
# Create pandas dataframe 
X_train = pd.DataFrame(X_train, columns=all_features)
X_test = pd.DataFrame(X_test, columns=all_features)
```

# Build Model

```python
# build a Logitic Regression model

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_predictions = log_reg.predict(X_train)
```

```python
# measure accuracy
from sklearn.model_selection import cross_val_score
cross_val_score(log_reg, X_train, y_train, cv=3, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
log_reg_pred = cross_val_predict(log_reg, X_train, y_train, cv=3)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, log_reg_pred)
```

```python
final_predictions = log_reg.predict(X_test)
confusion_matrix(y_test, final_predictions)
```

The accuracy on the test set is 100% as we can see there is no error.
